```python
%pylab inline
#load the list of points
obs_data = loadtxt("traces.dat")
x_obs = np.array(obs_data[:,0])
y_obs = np.array(obs_data[:,1])
x2_obs = np.array(obs_data[:,3])
y2_obs = np.array(obs_data[:,4])
z_obs = np.array(obs_data[:,2])
z2_obs = np.array(obs_data[:,5])
```

```python
def likelihood(y, y_model, x, x_model,):
    chi_squared = sum(((y-y_model)**2)+((x-x_model)**2))
    return (-chi_squared)
```

```python
def my_modely(y1, z1, y, z, z2):
    return ((y1 - y)/(z1- z))*z2 + y1
```

```python
def my_modelx(x1, z1, x, z,z2):
    return ((x1 - x)/(z1- z))*z2 + x1
```

```python
x_walk = empty((0))
y_walk = empty((0))
z_walk = empty((0))

x_walk = append(x_walk, random.random())
y_walk = append(y_walk, random.random())
z_walk = -2000.0*append(z_walk, random.random())

print x_walk
print y_walk
print z_walk
```

```python
n_iterations = 10000 #this is the number of iterations I want to make
cont=0
cont2=0

for i in range(n_iterations):
        
    x0_prime = np.random.normal(x_walk[i], 1.0)
    y0_prime = np.random.normal(y_walk[i], 1.0) 
    z0_prime = np.random.normal(z_walk[i], 100.0)
    while (z0_prime >= 0 or z0_prime <= -5000.0):	
        z0_prime = numpy.random.normal(z_walk[i], 200.0)



    y_init = my_modely(y_obs,z_obs ,y_walk[i], z_walk[i], z2_obs)
    x_init = my_modelx(x_obs,z_obs ,x_walk[i], z_walk[i], z2_obs)
    y_prime = my_modely(y_obs,z_obs ,y0_prime, z0_prime, z2_obs)
    x_prime = my_modelx(x_obs,z_obs ,x0_prime, z0_prime, z2_obs)
    
    
    alpha = likelihood(y2_obs, y_prime, x2_obs, x_prime )/likelihood(y2_obs, y_init, x2_obs, x_init)


    if(alpha<=1.0):
        z_walk  = append(z_walk,z0_prime)
        x_walk  = append(x_walk,x0_prime)
        y_walk  = append(y_walk,y0_prime)
        cont+=1
       
    else:
        beta = random.random()
        if(beta<=exp(-alpha)):
            z_walk  = append(z_walk,z0_prime)
            x_walk  = append(x_walk,x0_prime)
            y_walk  = append(y_walk,y0_prime)
            cont2+=1
            
        else:
            z_walk  = append(z_walk,z_walk[i])
            x_walk  = append(x_walk,x_walk[i])
            y_walk  = append(y_walk,y_walk[i])
            cont2+=1

    

            
           
```

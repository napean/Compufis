```python
x0_particulas = empty ((0))
y0_particulas = empty ((0))
vx0_particulas = empty ((0))
vy0_particulas = empty ((0))
parse = 3E16 
G = 6.7E-11
M = 2E42 
```

```python
def funx_prim (vx):
    return vx

def funy_prim (  vy):
    return vy

def funx_prim2( x ):
    return  -(6.7E-11*2E42)/(x**2)
    
def funy_prim2(  y ):    
    return  -(6.7E-11*2E42)/(y**2)
    
```

```python

h = 1000
n_points = (2000000)/h


momento1 = [[0 for x in xrange(4)]for x in xrange(100)]
momento2  = [[0 for x in xrange(4)]for x in xrange(100)]
momento3 = [[0 for x in xrange(4)]for x in xrange(100)]
momento4 = [[0 for x in xrange(4)]for x in xrange(100)]

w=0

while w < 100:
    print w
    posx = empty(n_points)
    posx[0] = x0_particulas[w]
    posy = empty(n_points)
    posy[0] = y0_particulas[w] 
    vx = empty (n_points)
    vx[0] = vx0_particulas[w] 
    vy = empty (n_points)
    vy[0] =  vy0_particulas[w] 
    
  

    
    for pp in range(1, n_points):
  

        
        k1_x = funx_prim(vx[pp-1])
        k1_y = funy_prim(vy[pp-1])
        k1_vx = funx_prim2(posx[pp-1])
        k1_vy = funy_prim2(posy[pp-1])
    
        #first step
      
        posx1 = posx[pp-1] + (h/2.0)*k1_x
        posy1 = posy[pp-1] + (h/2.0) * k1_y
        vx1 = vx[pp-1] + (h/2.0) * k1_vx
        vy1 = vy[pp-1] + (h/2.0 )* k1_vy
        
        k2_x = funx_prim(vx1)
        k2_y = funy_prim(vy1)
        k2_vx = funx_prim2(posx1)
        k2_vy = funy_prim2(posy1)
        
        #second step
        posx2 = posx[pp-1] + (h/2.0)*k2_x
        posy2 = posy[pp-1] + (h/2.0) * k2_y
        vx2 = vx[pp-1] + (h/2.0)*k2_vx
        vy2 = vy[pp-1] + (h/2.0)*k2_vy
        
        k3_x = funx_prim(vx2)
        k3_y = funy_prim(vy2)
        k3_vx = funx_prim2(posx2)
        k3_vy = funy_prim2(posy2)
        
        #third step
        
        posx3 = posx[pp-1] + (h/2.0)*k3_x
        posy3 = posy[pp-1] + (h/2.0) * k3_y
        vx3 = vx[pp-1] + (h/2.0)*k3_vx
        vy3 = vy[pp-1] + (h/2.0)*k3_vy
        
        k4_x = funx_prim(vx3)
        k4_y = funy_prim(vy3)
        k4_vx = funx_prim2(posx3)
        k4_vy = funy_prim2(posy3)
        
        #fourth step
        average_kx = (1.0/6.0)*(k1_x + 2.0*k2_x + 2.0*k3_x + k4_x)
        average_ky = (1.0/6.0)*(k1_y + 2.0*k2_y + 2.0*k3_y + k4_y)
        average_kvx = (1.0/6.0)*(k1_vx + 2.0*k2_vx + 2.0*k3_vx + k4_vx)
        average_kvy = (1.0/6.0)*(k1_vy + 2.0*k2_vy + 2.0*k3_vy + k4_vy)
        
        posx[pp] =  posx[pp-1] + h* average_kx
        posy[pp] =  posy[pp-1] + h* average_ky
        posx[pp] =  vx[pp-1] + h* average_kvx
        posx[pp] =  vy[pp-1] + h* average_kvy
        

        
    w+= 1;
```
